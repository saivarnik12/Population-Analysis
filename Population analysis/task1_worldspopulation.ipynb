```python
import pandas as pd

# Read the CSV file, skipping the first 3 lines
df = pd.read_csv('world_population.csv', delimiter=',', encoding='utf-8', skiprows=2)

# Print the DataFrame
print(df)
```

```python
import pandas as pd

df.fillna(0, inplace=True)
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Extracting years
years = df.columns[5:]

# List of countries you want to plot
country_names = ['Aruba', 'Africa Eastern and Southern', 'Afghanistan', 'Africa Western and Central',
                 'Angola', 'Albania', 'Andorra', 'Argentina', 'Australia', 'Austria', 'Azerbaijan']

# Plotting bar graph for each country
plt.figure(figsize=(18, 9))
for i, country in enumerate(country_names, 1):
    country_data = df[df['Country Name'] == country].iloc[:, 5:].values.flatten()
    plt.subplot(3, 4, i)
    plt.bar(years, country_data, label=country)
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title(f'Population of {country} Over Years')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.gca().margins(x=0.05)
    plt.gca().tick_params(axis='x', which='major', pad=5)
    plt.gca().set_xticks(years[::5]) # with the interval of 5 years

# for  subplot spacing
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt



# column after "indicator Name "
years = df.columns[4:]

# Summing up population data for each year across all countries
total_population = []
for year in years:
    total_population.append(df[year].sum())

# Plotting bar graph for total population over the years
plt.figure(figsize=(12, 6))
plt.bar(years, total_population, color='blue')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total Population Over Years')
plt.xticks(rotation=45)  # for better readability
plt.grid(True)
plt.show()
```
