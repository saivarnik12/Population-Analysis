```python


# Read the CSV file, skipping the first 3 lines
df = pd.read_csv('country_population.csv')

# Print the DataFrame
print(df)


```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Extract relevant columns
df_subset = df[['IncomeGroup', 'TableName']]

# Drop NaN values
df_subset = df_subset.dropna()

# Group by 'IncomeGroup' and count the occurrences of each income group
grouped_data = df_subset.groupby('IncomeGroup').count()

# Plotting bar graph
plt.figure(figsize=(10, 6))
plt.bar(grouped_data.index, grouped_data['TableName'])
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Count of Countries by Income Group')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(df_subset['IncomeGroup'], bins=len(grouped_data), edgecolor='black')
plt.xlabel('Income Group')
plt.ylabel('Frequency')
plt.title('Histogram of Income Groups')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the count plot for Region and IncomeGroup
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Region', hue='IncomeGroup')
plt.title('Count of Income Groups in Each Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income Group')
plt.show()
```
